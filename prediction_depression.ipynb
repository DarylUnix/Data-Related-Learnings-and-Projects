# **DEPRESSION DATASET DESCRIPTION**
We all know now a days mental health is one of the major problem among the students in the current days, So the dataset of some students is provided such that anyone can make health detection system that the student has any such type of depression or not. Depression.csv is basically the training file and the test.csv is for the testing the model with true labels.

Kaggle Link: https://www.kaggle.com/datasets/indubarnwal/depression-dataset

# **DATA PREPARATION**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, classification_report

from google.colab import drive
drive.mount("/content/drive/", force_remount=True)
df = pd.read_csv("/content/drive/MyDrive/Dataset/Depression.csv", low_memory=False)
df
```

```python
df.info()
```

# **DATA CLEANING**

```python
check = df.isnull().sum()
print("Missing values found? F/N: ", check.all())
```

```python
df['T_Ignored'] = df['Ignored'].replace({'Sleep quality': 0,'Diet': 1, 'Exercise': 2})
df['T_Gender'] = df['Gender'].replace({'Male': 0, 'Female': 1})
df['T_Demographics'] = df['Demographics'].replace({'Rural': 0, 'Urban': 1})
df['T_Family_Problem'] = df['Family Problem'].replace({'No': 0, 'Yes': 1})
df['T_Social_Support'] = df['Social Support'].replace({'Low': 0, 'Moderate': 1, 'High':2})
df['T_CGPA'] = df['CGPA'].replace({'Moderate': 1, 'Low': 0, 'High': 2})
df['T_Financial_Problem'] = df['Financial Problem'].replace({'Mild': 0, 'Moderate': 1, 'Significant':2})
print(df.info())
```

```python
df.drop('Gender', axis=1, inplace=True)
df.drop('Ignored', axis=1, inplace=True)
df.drop('Family Problem', axis=1, inplace=True)
df.drop('Demographics', axis=1, inplace=True)
df.drop('Social Support', axis=1, inplace=True)
df.drop('CGPA', axis=1, inplace=True)
df.drop('Financial Problem', axis=1, inplace=True)
df.info()
```

# **SPLITTING**

```python
x = df.drop('Depression', axis=1)
x
```

```python
y = df['Depression']
y.values
```

```python
x_train, x_test, y_train, y_test=train_test_split (x,y,test_size=0.2)
```

```python
x_train
```

```python
x_test
```

```python
y_train.values
```

# **BUILD MODEL**

```python
model = GaussianNB()
model.fit(x_train, y_train)
```

```python
predicted_value=model.predict(x_test)
predicted_value
```

```python
tp=0
tn=0
fn=0
fp=0

featured_value = np.array(x_test.values)
real = np.array(y_test.values)
predicted = np.array(predicted_value)
age = np.array(['Age'])

for i in range (y_test.size):
  if  predicted[i] == real[i]:
    classifier = "True Positive"
    tp=tp+1
  if  predicted[i] == real[i]==0:
    classifier = "True Negative"
    tn=tn+1
  if  predicted[i]==0 and real[i]==1:
    classifier = "False Positive"
    fp=fp+1
  if  predicted[i]==1 and real[i]==0:
    classifier = "False Negative"
    fn=fn+1

print("True Positive (TP): ", tp)
print("True Negative (TN): ", tn)
print("False Positive (FP): ", fp)
print("False Negative (FN): ", fn)
```

# **CONFUSION MATRIX**

```python
cm = confusion_matrix(predicted_value, y_test)
cm
```

```python
accuracy_score(predicted_value, y_test)*100
```

# **TEST CASE**

```python
NOC = int(input('Please Input Number of cases: '))
for x in range(1, NOC + 1):
    print(f"Case {x}\n")

    df_sample = pd.DataFrame(columns=['Age','T_Ignored','T_Gender','T_Demographics','T_Family_Problem', 'T_Social_Support', 'T_CGPA', 'T_Financial_Problem'])
    while True:
      g_answer = input('Input your Gender (Male/Female): ')
      if g_answer == "Male":
        T_Gender = 0
        break
      elif g_answer == "Female":
        T_Gender = 1
        break
      else:
        print("Input is wrong! Please try again")
    while True:
      d_answer = input('Input your Area Demographic (Rural/Urban): ')
      if d_answer == "Rural":
        T_Demographics = 0
        break
      elif d_answer == "Urban":
        T_Demographics = 1
        break
      else:
        print("Input is wrong! Please try again")
    while True:
      F_ans = input('Are you currently experiencing family problems? (Yes/No): ')
      if F_ans == "No":
        T_Family_Problem = 0
        break
      elif F_ans == "Yes":
        T_Family_Problem = 1
        break
      else:
        print("Input is wrong! Please try again")
    while True:
      a_ans = input('Enter your Age (18-65): ')
      if int(a_ans) >= 18 and int(a_ans) <= 65:
        Age = a_ans
        break
      else:
        print("Input is not in Age Range! Please try again")
    while True:
      i_ans = input('Which of these factors have you been ignoring lately (Sleep/Diet/Exercise): ')
      if i_ans == "Sleep":
        T_Ignored = 0
        break
      elif i_ans == "Diet":
        T_Ignored = 1
        break
      elif i_ans == "Exercise":
        T_Ignored = 2
        break
      else:
        print("Input is wrong! Please try again")
    while True:
      ss_ans = input('Do you receive any form of social support from friends, peers, or family? (Low/Moderate/High): ')
      if ss_ans == "Low":
        T_Social_Support = 0
        break
      elif ss_ans == "Moderate":
        T_Social_Support = 1
        break
      elif ss_ans == "High":
        T_Social_Support = 2
        break
      else:
        print("Input is wrong! Please try again")
    while True:
      c_ans = input ('What is your CGPA level? (Low/Average/High): ')
      if c_ans == "Low":
        T_CGPA = 0
        break
      elif c_ans == "Average":
        T_CGPA = 1
        break
      elif c_ans == "High":
        T_CGPA = 2
        break
      else:
        print("Input is wrong! Please try again")
    while True:
      fp_ans = input('Are you currently experiencing any financial problems? (Mild/Average/Significant): ')
      if fp_ans == "Mild":
        T_Financial_Problem = 0
        break
      elif fp_ans == "Average":
        T_Financial_Problem = 1
        break
      elif fp_ans == "Significant":
        T_Financial_Problem = 2
        break
      else:
        print("Input is wrong! Please try again")

    df_sample.loc[len(df_sample.index)]=[Age,T_Ignored,T_Gender,T_Demographics,T_Family_Problem,T_Social_Support,T_CGPA,T_Financial_Problem]

    pred = model.predict(df_sample)
    if pred == 1:
      print("\nIt is possible that you are currently experiencing symptoms of depression. It's important to seek professional help if you can. Here are some additional recommendations to consider:\n"
          "1. **Seek Professional Help**: Consult with a mental health professional such as a therapist or counselor.\n"
          "2. **Support Groups**: Consider joining a support group where you can share experiences and receive support.\n"
          "3. **Healthy Lifestyle**: Maintain a balanced diet and engage in regular physical activity.\n"
          "4. **Avoid Substances**: Steer clear of alcohol and drugs, as they can exacerbate symptoms.\n"
          "5. **Relaxation Techniques**: Practice relaxation techniques such as meditation, deep breathing exercises, or yoga.\n"
          "6. **Social Support**: Reach out to trusted friends or family members for support and connection.\n")
    else:
      print("\nYou are not currently showing signs of depression. Continue to maintain a healthy lifestyle. Here are some tips to stay well:\n"
          "1. **Stay Active**: Engage in regular exercise to boost your mood and energy levels.\n"
          "2. **Balanced Diet**: Eat a balanced and nutritious diet to support your overall well-being.\n"
          "3. **Stay Connected**: Keep in touch with friends and family to maintain strong social connections.\n"
          "4. **Relaxation Techniques**: Incorporate relaxation practices such as meditation or yoga into your routine.\n"
          "5. **Seek Help if Needed**: If you ever feel overwhelmed, don't hesitate to seek professional help.\n")
```
