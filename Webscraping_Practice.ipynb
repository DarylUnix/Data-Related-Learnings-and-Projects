```python
from bs4 import BeautifulSoup
```

```python
import requests

url = 'https://basketball.realgm.com/nba/summer/1/NBA-Summer-League/15/stats'

page = requests.get(url)  # Call the get method on the requests module

page
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
all_headers = soup.find_all('a')
all_headers
```

```python
elements = soup.find_all('a', href=lambda href: href and "/nba/summer/1/NBA-Summer-League/15/stats/NBA/0/Averages/All" in href)
elements
```

```python
headers = [element.text for element in elements]
headers
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns=headers)
df.insert(0,"ID","")
df.insert(2,"Team","")
df
```

```python
column_data = soup.find_all('td')
column_data
```

```python
all_rows = []
rows = []
for element in column_data:
    rows.append(element.text)
    if len(rows) == len(df.columns):
        all_rows.append(rows)
        rows = []

all_rows
```

```python
df = pd.DataFrame(all_rows, columns=df.columns)
df
```
